```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("sales_data_sample.csv", encoding="latin1")
```

```python
data
```

```python
data.isnull().sum()
```

```python
# Prepare the data for clustering (select relevant columns and handle missing values)
selected_data = data[['ORDERNUMBER', 'QTR_ID', 'MONTH_ID']]
selected_data = selected_data.dropna()
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(selected_data)
```

```python
wcss = []  # Within-Cluster-Sum-of-Squares

for i in range(1, 11):  # You can change the range based on your problem
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(data_scaled)
    wcss.append(kmeans.inertia_)  # Inertia: WCSS
```

```python
# Plot the elbow graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Choose the optimal number of clusters (based on the elbow plot)
optimal_clusters = 3  # Adjust this based on your elbow plot
# Perform K-Means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(data_scaled)
print("Cluster Labels:")
print(cluster_labels)
```
